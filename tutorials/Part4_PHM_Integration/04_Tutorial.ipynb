```python
# 3. 信号处理算子演示
print("📈 信号处理算子演示")

if mode == 'full_integration':
    try:
        # 真实信号处理算子
        print(f"\n🔧 可用信号处理算子: {len(OP_REGISTRY)}个")
        
        # 按类别展示算子
        categories = {}
        for op_name, op_info in OP_REGISTRY.items():
            category = op_info.get('category', 'Other')
            if category not in categories:
                categories[category] = []
            categories[category].append(op_name)
        
        for category, operators in categories.items():
            print(f"\n{category}算子 ({len(operators)}个):")
            for op in operators[:5]:  # 显示前5个
                print(f"  - {op}")
            if len(operators) > 5:
                print(f"  ... 还有{len(operators)-5}个")
        
        # 演示算子使用
        print("\n⚡ 算子使用演示:")
        
        # 获取FFT算子
        from src.tools.signal_processing_schemas import get_operator
        
        if 'fft' in OP_REGISTRY:
            fft_op = get_operator('fft')()
            print(f"  ✅ FFT算子加载成功: {type(fft_op)}")
        
        if 'rms' in OP_REGISTRY:
            rms_op = get_operator('rms')()
            print(f"  ✅ RMS算子加载成功: {type(rms_op)}")
            
    except Exception as e:
        print(f"⚠️ 真实算子演示失败: {e}")
        mode = 'simulation'

if mode in ['demo_mode', 'simulation']:
    print("\n🔄 模拟信号处理算子演示:")
    
    # 模拟算子分类
    operator_categories = {
        "时域分析": ["RMS", "峰值", "峭度", "偏度", "能量"],
        "频域分析": ["FFT", "PSD", "频谱质心", "频带能量", "谐波分析"],
        "时频分析": ["小波变换", "短时傅里叶", "希尔伯特变换"],
        "滤波处理": ["低通滤波", "高通滤波", "带通滤波", "陷波滤波"],
        "特征提取": ["包络分析", "倒频谱", "相关分析", "熵特征"],
        "预处理": ["去均值", "归一化", "重采样", "去趋势"]
    }
    
    total_operators = sum(len(ops) for ops in operator_categories.values())
    print(f"\n📊 信号处理算子生态 (共{total_operators}个):")
    
    for category, operators in operator_categories.items():
        print(f"\n{category} ({len(operators)}个):")
        for op in operators:
            print(f"  ✅ {op}")
    
    # 模拟处理流程
    print("\n⚡ 模拟信号处理流程:")
    processing_flow = [
        "原始信号 → 去均值 → 低通滤波 → FFT → 频谱分析",
        "振动信号 → RMS计算 → 峭度分析 → 包络检波 → 故障特征", 
        "多通道信号 → 相关分析 → 主成分分析 → 特征融合"
    ]
    
    for i, flow in enumerate(processing_flow, 1):
        print(f"  {i}. {flow}")
```
