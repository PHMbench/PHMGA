```python
# 3. Routeræ¨¡å¼æ¼”ç¤º
print("ğŸ”€ Routeræ¨¡å¼æ¼”ç¤º")

try:
    # è§„åˆ™åŸºç¡€Router
    print("\n1ï¸âƒ£ è§„åˆ™åŸºç¡€Router:")
    rule_router = router_patterns.RuleBasedRouter()
    
    test_requests = [
        {"sensor_type": "temperature", "value": 95, "urgency": "high"},
        {"sensor_type": "vibration", "value": 3.2, "urgency": "low"},
        {"sensor_type": "pressure", "value": 0.8, "urgency": "medium"}
    ]
    
    for req in test_requests:
        route = rule_router.route(req)
        print(f"è¯·æ±‚: {req['sensor_type']} â†’ è·¯ç”±: {route}")
    
    # LLMåŸºç¡€Router
    print("\n2ï¸âƒ£ LLMæ™ºèƒ½Router:")
    llm_router = router_patterns.LLMBasedRouter()
    
    complex_request = {
        "description": "è®¾å¤‡å‡ºç°é—´æ­‡æ€§å¼‚å“ï¼Œæ¸©åº¦ç•¥é«˜äºæ­£å¸¸å€¼",
        "context": "æœ€è¿‘è¿›è¡Œè¿‡ç»´æŠ¤ï¼Œä½†é—®é¢˜ä»ç„¶å­˜åœ¨"
    }
    
    intelligent_route = llm_router.route(complex_request)
    print(f"å¤æ‚è¯·æ±‚ â†’ æ™ºèƒ½è·¯ç”±: {intelligent_route}")
    
    # è´Ÿè½½å‡è¡¡Router
    print("\n3ï¸âƒ£ è´Ÿè½½å‡è¡¡Router:")
    lb_router = router_patterns.LoadBalancingRouter()
    
    for i in range(5):
        route = lb_router.route({"request_id": i})
        print(f"è¯·æ±‚{i} â†’ åˆ†é…åˆ°: {route}")
        
except Exception as e:
    print(f"âš ï¸ Routeræ¼”ç¤ºå¤±è´¥: {e}")
    print("ğŸ”„ ä½¿ç”¨æ¨¡æ‹Ÿæ¼”ç¤º:")
    print("è§„åˆ™Router: temperatureâ†’thermal_analysis, vibrationâ†’mechanical_analysis")
    print("LLM Router: æ™ºèƒ½åˆ†æè¯·æ±‚è¯­ä¹‰ï¼Œé€‰æ‹©æœ€ä¼˜å¤„ç†è·¯å¾„")
    print("è´Ÿè½½å‡è¡¡: è¯·æ±‚åœ¨å¤šä¸ªå¤„ç†èŠ‚ç‚¹é—´å‡åŒ€åˆ†é…")
```

```python
# 4. è‡ªé€‚åº”Routeræ¼”ç¤º
print("ğŸ§  è‡ªé€‚åº”Routeræ¼”ç¤º")

try:
    # åˆ›å»ºè‡ªé€‚åº”Router
    adaptive_router = router_patterns.AdaptiveRouter()
    
    # æ¨¡æ‹Ÿä¸€ç³»åˆ—è¯·æ±‚å’Œåé¦ˆ
    test_scenarios = [
        {"sensor_data": {"temperature": 90, "vibration": 8.0}, "priority": "high"},
        {"sensor_data": {"temperature": 70, "vibration": 2.1}, "priority": "low"},
        {"sensor_data": {"temperature": 85, "vibration": 6.5}, "priority": "medium"}
    ]
    
    print("\nè‡ªé€‚åº”è·¯ç”±å†³ç­–:")
    for i, scenario in enumerate(test_scenarios):
        route_decision = adaptive_router.route(scenario)
        print(f"åœºæ™¯{i+1}: {scenario['priority']}ä¼˜å…ˆçº§ â†’ {route_decision}")
        
        # æ¨¡æ‹Ÿæ€§èƒ½åé¦ˆ
        performance_score = 0.8 if route_decision.startswith('fast') else 0.9
        adaptive_router.update_performance(route_decision, performance_score)
    
    # æ˜¾ç¤ºå­¦ä¹ æ•ˆæœ
    print("\nğŸ“ˆ Routerå­¦ä¹ ç»Ÿè®¡:")
    stats = adaptive_router.get_performance_stats()
    for route, performance in stats.items():
        print(f"  {route}: å¹³å‡æ€§èƒ½ {performance:.2f}")
        
except Exception as e:
    print(f"âš ï¸ è‡ªé€‚åº”Routeræ¼”ç¤ºå¤±è´¥: {e}")
    print("ğŸ”„ ä½¿ç”¨æ¨¡æ‹Ÿæ¼”ç¤º:")
    print("è‡ªé€‚åº”å­¦ä¹ : Routeræ ¹æ®å†å²æ€§èƒ½è°ƒæ•´è·¯ç”±ç­–ç•¥")
    print("æ€§èƒ½åé¦ˆ: æ”¶é›†å¤„ç†ç»“æœï¼Œä¼˜åŒ–åç»­å†³ç­–")
    print("åŠ¨æ€è°ƒæ•´: å®æ—¶è°ƒæ•´è·¯ç”±æƒé‡å’Œä¼˜å…ˆçº§")
```

```python
# 5. ç»„åˆç³»ç»Ÿæ¼”ç¤º
print("ğŸ”— ç»„åˆç³»ç»Ÿæ¼”ç¤º")

try:
    # åˆ›å»ºç»„åˆçš„PHMå¤„ç†ç³»ç»Ÿ
    print("\næ„å»ºå®Œæ•´çš„PHMå¤„ç†æµæ°´çº¿:")
    
    # 1. çŠ¶æ€ç®¡ç†å™¨
    state_manager = state_management.StateManager()
    print("âœ“ çŠ¶æ€ç®¡ç†å™¨å·²åˆ›å»º")
    
    # 2. æ™ºèƒ½è·¯ç”±å™¨
    smart_router = router_patterns.LLMBasedRouter()
    print("âœ“ æ™ºèƒ½è·¯ç”±å™¨å·²åˆ›å»º")
    
    # 3. å·¥ä½œæµå¼•æ“
    workflow_engine = langgraph_workflows.AdvancedPHMWorkflow()
    print("âœ“ å·¥ä½œæµå¼•æ“å·²åˆ›å»º")
    
    # æ¨¡æ‹Ÿå®Œæ•´å¤„ç†æµç¨‹
    input_data = {
        "equipment_id": "MOTOR-002",
        "sensor_readings": {"temperature": 88.5, "vibration": 7.2, "current": 15.8},
        "alert_description": "ç”µæœºè¿è¡Œæ—¶å‡ºç°ä¸è§„å¾‹å™ªéŸ³"
    }
    
    print(f"\nè¾“å…¥æ•°æ®: {input_data['equipment_id']}")
    
    # Step 1: è·¯ç”±å†³ç­–
    route = smart_router.route(input_data)
    print(f"1. è·¯ç”±å†³ç­–: {route}")
    
    # Step 2: çŠ¶æ€åˆå§‹åŒ–
    initial_state = state_management.create_phm_diagnosis_state(
        input_data["equipment_id"], 
        input_data["sensor_readings"]
    )
    state_manager.set_state(initial_state)
    print("2. çŠ¶æ€åˆå§‹åŒ–å®Œæˆ")
    
    # Step 3: å·¥ä½œæµæ‰§è¡Œ
    analysis_result = workflow_engine.run_analysis(input_data["sensor_readings"])
    print(f"3. åˆ†æå®Œæˆ: {analysis_result.get('comprehensive_diagnosis', 'N/A')[:50]}...")
    
    # Step 4: çŠ¶æ€æ›´æ–°
    final_update = {
        "current_diagnosis": analysis_result.get('comprehensive_diagnosis', 'æœªçŸ¥'),
        "severity_level": "high" if "critical" in str(analysis_result).lower() else "medium",
        "recommended_actions": ["ç«‹å³æ£€æŸ¥", "è®¡åˆ’ç»´æŠ¤"]
    }
    final_state = state_manager.update_state(final_update)
    print("4. æœ€ç»ˆçŠ¶æ€æ›´æ–°å®Œæˆ")
    
    print("\nğŸ‰ å®Œæ•´PHMå¤„ç†æµæ°´çº¿æ¼”ç¤ºæˆåŠŸï¼")
    
except Exception as e:
    print(f"âš ï¸ ç»„åˆç³»ç»Ÿæ¼”ç¤ºå¤±è´¥: {e}")
    print("ğŸ”„ æ¨¡æ‹Ÿç»„åˆæµç¨‹:")
    print("è¾“å…¥ â†’ æ™ºèƒ½è·¯ç”± â†’ çŠ¶æ€ç®¡ç† â†’ å·¥ä½œæµæ‰§è¡Œ â†’ ç»“æœè¾“å‡º")
    print("âœ“ æ‰€æœ‰ç»„ä»¶æ— ç¼é›†æˆ")
    print("âœ“ çŠ¶æ€åœ¨æ•´ä¸ªæµç¨‹ä¸­ä¿æŒä¸€è‡´")
    print("âœ“ æ”¯æŒå¤æ‚çš„å¤šæ­¥éª¤åˆ†æ")
```
