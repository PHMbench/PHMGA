## Step 2: å®‰è£…æ ¸å¿ƒä¾èµ–

```python
import subprocess
import sys

def install_package(package):
    """å®‰è£…PythonåŒ…"""
    try:
        subprocess.check_call([sys.executable, "-m", "pip", "install", package])
        return True
    except subprocess.CalledProcessError:
        return False

# æ ¸å¿ƒä¾èµ–åŒ…
core_packages = [
    "langchain",
    "langchain-core", 
    "langgraph",
    "langchain-community",
    "numpy",
    "scipy",
    "matplotlib",
    "pandas",
    "ipywidgets",
    "pydantic"
]

print("ğŸ“¦ å¼€å§‹å®‰è£…æ ¸å¿ƒä¾èµ–åŒ…...")
for package in core_packages:
    print(f"å®‰è£… {package}...", end=" ")
    if install_package(package):
        print("âœ…")
    else:
        print("âŒ")
```

```python
from IPython.display import display, HTML

# ç”Ÿæˆé…ç½®æŠ¥å‘Š
def generate_config_report():
    """ç”Ÿæˆç¯å¢ƒé…ç½®æŠ¥å‘Š"""
    
    # æ£€æŸ¥Pythonç‰ˆæœ¬
    python_ok = sys.version_info >= (3, 8)
    
    # æ£€æŸ¥æ ¸å¿ƒåŒ…
    core_packages_status = {}
    for pkg in ["langchain", "langgraph", "numpy", "matplotlib"]:
        try:
            __import__(pkg.replace("-", "_"))
            core_packages_status[pkg] = True
        except ImportError:
            core_packages_status[pkg] = False
    
    # æ£€æŸ¥APIå¯†é’¥
    api_status = {}
    for key, name in api_keys:
        api_status[name] = bool(os.getenv(key, "").strip())
    
    # æ£€æŸ¥é¡¹ç›®å¯¼å…¥
    try:
        from src.model.providers import get_provider_registry
        project_import_ok = True
    except ImportError:
        project_import_ok = False
    
    return {
        "python_ok": python_ok,
        "core_packages": core_packages_status,
        "api_keys": api_status,
        "project_import": project_import_ok
    }

report = generate_config_report()

# æ˜¾ç¤ºæŠ¥å‘Š
html_report = f"""
<div style="background-color: #f8f9fa; border-radius: 10px; padding: 20px; margin: 20px 0;">
    <h2 style="color: #2c3e50; margin-top: 0;">ğŸ¯ ç¯å¢ƒé…ç½®æ£€æŸ¥æŠ¥å‘Š</h2>
    
    <h3>âœ… åŸºç¡€ç¯å¢ƒ</h3>
    <ul>
        <li>Python 3.8+: {'âœ… é€šè¿‡' if report['python_ok'] else 'âŒ éœ€è¦å‡çº§'}</li>
        <li>é¡¹ç›®å¯¼å…¥: {'âœ… æ­£å¸¸' if report['project_import'] else 'âŒ å¤±è´¥'}</li>
    </ul>
    
    <h3>ğŸ“¦ æ ¸å¿ƒä¾èµ–</h3>
    <ul>
"""

for pkg, status in report['core_packages'].items():
    html_report += f"<li>{pkg}: {'âœ… å·²å®‰è£…' if status else 'âŒ æœªå®‰è£…'}</li>"

html_report += """
    </ul>
    
    <h3>ğŸ”‘ APIå¯†é’¥</h3>
    <ul>
"""

for provider, status in report['api_keys'].items():
    html_report += f"<li>{provider}: {'âœ… å·²é…ç½®' if status else 'âŒ æœªé…ç½®'}</li>"

configured_count = sum(report['api_keys'].values())
overall_status = (
    report['python_ok'] and 
    report['project_import'] and 
    all(report['core_packages'].values()) and
    (configured_count > 0)  # è‡³å°‘ä¸€ä¸ªAPIå¯†é’¥æˆ–è€…å¯ä»¥ç”¨Mock
)

status_color = "#27ae60" if overall_status else "#e74c3c"
status_text = "ğŸ‰ ç¯å¢ƒé…ç½®å®Œæˆï¼" if overall_status else "âš ï¸ éœ€è¦å®Œå–„é…ç½®"

html_report += f"""
    </ul>
    
    <div style="background-color: {status_color}; color: white; padding: 15px; border-radius: 5px; margin-top: 20px;">
        <h3 style="margin: 0;">{status_text}</h3>
        {'<p>æ‚¨å·²ç»å‡†å¤‡å¥½å¼€å§‹PHMGAæ•™ç¨‹äº†ï¼</p>' if overall_status else '<p>è¯·å®Œå–„ä¸Šè¿°é…ç½®åå†å¼€å§‹æ•™ç¨‹ã€‚</p>'}
    </div>
</div>
"""

display(HTML(html_report))
```
